```python
def split_into_high_low(date,kalshi_mid):
    strike_low = kalshi_mid - 12
    strike_high = kalshi_mid + 13

    file = pd.read_csv('Raw Data/UnderlyingOptionsIntervals_60sec_oi_'+date+'.csv')
    filtered_data = file[file['expiration'] == date]

    low_data = filtered_data[filtered_data['strike'] == strike_low]
    high_data = filtered_data[filtered_data['strike'] == strike_high]
    
    return low_data,high_data
```

```python
def create_cleaned_data(data):
    data['quote_datetime'] = pd.to_datetime(data['quote_datetime'])
    spx_data['Date Time'] = pd.to_datetime(spx_data['Date Time'])

    # You would need to ensure the time is rounded to the nearest minute in csv_df if it's not already
    spx_data['Date Time'] = spx_data['Date Time'].dt.floor('T')
    data['quote_datetime'] = data['quote_datetime'].dt.floor('T')

    # Merge the DataFrames based on the rounded datetime
    merged_df = data.merge(spx_data, left_on='quote_datetime', right_on='Date Time', how='left')
    
    def select_ask(row):
        if row['strike'] > row['Open']:
            if row['option_type'] == 'C':
                return row['ask']  # Use the ask price from the Call option
        else:
            if row['option_type'] == 'P':
                return row['ask']  # Use the ask price from the Put option
        return None  # In case there is no matching condition

    # Apply the function to each row to create a new 'selected_ask' column
    merged_df['selected_ask'] = merged_df.apply(select_ask, axis=1)

    # Filter the DataFrame to rows where 'selected_ask' is not None
    result_df = merged_df[merged_df['selected_ask'].notnull()]
    
    final_df = result_df[['quote_datetime','ask','option_type']]
    final_df['quote_datetime'] = final_df['quote_datetime'].dt.strftime('%H:%M')



    return final_df
```

```python
import pandas as pd
import os
from datetime import datetime
import calendar


def store_dataframe(final_df, date_string, kalshi_mid, is_low):
    
    # Assuming you have a DataFrame named df

    # Parse the date string to a datetime object
    date_object = datetime.strptime(date_string, '%Y-%m-%d')

    # Extract year, month name, and day from the date object
    year = date_object.strftime('%y')  # Two-digit year
    month_name = calendar.month_name[date_object.month]  # Full month name
    day = date_object.strftime('%d')  # Two-digit day

    # Create a folder path in the format 'YY/MONTH_NAME/DD'
    folder_path = os.path.join(year, month_name, day)

    # Check if the directory exists, if not, create it
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)

    if is_low:
        file_name = 'Options-'+year+month_name+day+'-L'+str(kalshi_mid)
    else:
        file_name = 'Options-'+year+month_name+day+'-H'+str(kalshi_mid)

    # Now save the DataFrame in this directory
    # Replace 'your_dataframe.csv' with your actual DataFrame name
    file_path = os.path.join(folder_path, file_name+'.csv')
    final_df.to_csv(file_path, index=False)

    print(f"DataFrame saved in: {file_path}")
```
